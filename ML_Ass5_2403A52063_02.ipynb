```python
import pandas as pd

train_df = pd.read_csv(extract_path + "/train.csv")

train_df.head()
```

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
target = 'Survived'

X = train_df[features]
y = train_df[target]
```

```python
# Fill missing values
X['Age'] = X['Age'].fillna(X['Age'].median())
X['Embarked'] = X['Embarked'].fillna(X['Embarked'].mode()[0])

# Encode categorical variables
X = pd.get_dummies(X, columns=['Sex', 'Embarked'], drop_first=True)

X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
def train_and_evaluate(test_size):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42
    )

    model = LogisticRegression(max_iter=1000)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print(f"Train/Test Split: {int((1-test_size)*100)}/{int(test_size*100)}")
    print("Accuracy:", accuracy)
    print("Confusion Matrix:\n", cm)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1)
    print("-" * 50)

    return accuracy, precision, recall, f1
```

```python
results = {}

results['70/30'] = train_and_evaluate(0.30)
results['80/20'] = train_and_evaluate(0.20)
results['90/10'] = train_and_evaluate(0.10)
```

```python
results_df = pd.DataFrame(results,
                          index=['Accuracy', 'Precision', 'Recall', 'F1-score'])

results_df
```

From the experimental results, Logistic Regression was trained using
different train/test splits (70/30, 80/20, and 90/10).
The model with the 80/20 or 90/10 split generally shows better performance
due to more training data availability.
However, very small test sets may reduce reliability.
Hence, the 80/20 split provides a balanced and stable performance.
